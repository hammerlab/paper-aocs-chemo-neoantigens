```python
import pandas
```

```python
original = pandas.read_csv("../data/derived/annotated_mutations.missing_some_samples.csv.bz2")
```

```python
original.iloc[0]
```

```python
df = pandas.read_csv("../data/derived/annotated_mutations.csv.bz2")
df
```

```python
for some_df in [df, original]:
    for some_col in ["ref", "alt"]:
        some_df[some_col] = some_df[some_col].fillna("").map(lambda x: {"-": "", ".": ""}.get(x, x))
        
```

```python
DONORS_TO_EXCLUDE = [
    "AOCS-002",  # Not classified as HGSOC (email from Ann-Marie Patch)
]
```

```python
merged = df.merge(original[["donor", "contig", "interbase_start", "interbase_end", "ref", "alt", "context_mutation_3p5p", "binding_peptides", "num_binders"]],
                  on=("donor", "contig", "interbase_start", "interbase_end", "ref", "alt"),
                  how='right').drop_duplicates()
merged = merged.ix[~merged.donor.isin(DONORS_TO_EXCLUDE)]
merged.isnull().sum()
```

```python
bases = ["A", "C", "T", "G"]
```

```python
merged["indel"] = merged.ref.str.len() != merged.alt.str.len()
merged["snv"] = merged.ref.isin(bases) & merged.alt.isin(bases)
merged["mnv"] = (merged.ref.str.len() > 1) & (~merged.indel)
merged["confident"] = (merged.dna_alt_reads > 5) & (merged.dna_alt_reads / merged.dna_total_reads > .05)
merged["vaf"] = merged["dna_alt_reads"] / (merged["dna_total_reads"] + .0001)
merged["any_alt_reads"] = (merged.dna_alt_reads > 0) | (merged.rna_alt_reads > 0)
merged
```

```python
merged["mutation_id"] = [
        "%s:%d %s>%s" % (row.contig, row.interbase_start, row.ref, row.alt) for (i,row) in merged.iterrows()
    ]
merged["site_id"] = [
        "%s:%d" % (row.contig, row.interbase_start) for (i,row) in merged.iterrows()
    ]
```

```python
sources = pandas.read_csv("../data/derived/sources.full.csv", index_col="source_id")
sources["source_id"] = sources.index
```

```python
sites_in_primary = set(merged[merged.any_alt_reads][
        ~ merged[merged.any_alt_reads].source_id.isin(set(sources.index[sources.treated]))
    ].site_id)

sites_with_depth_in_primary = set(merged[
        (merged.dna_total_reads > 30) &
        (~ merged.source_id.isin(set(sources.index[sources.treated])))
    ].site_id)
```

```python
merged["unique_to_treated"] = (
        (merged.site_id.isin(sites_with_depth_in_primary)) &
        (~ merged.site_id.isin(sites_in_primary))
    )
```

```python
merged.to_csv("../data/derived/annotated_mutations.with_mhc_binders.csv", index=False)
!bzip2 -f ../data/derived/annotated_mutations.with_mhc_binders.csv
```
